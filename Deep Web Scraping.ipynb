```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url_card= "https://pacmann.io/blog/tag/artikel-data"
page= bs(requests.get(url_card).text)
```

```python
result=[]

all_card = page.findAll("div", class_="col-md-4")
for div in all_card[:-1]:
    # ambil dari masing-masing blogpost
    post_url = div.find("a")["href"]
    
    # extract single url 
    post= bs(requests.get(post_url).text)
    post_title = post.find("h1").get_text()
    post_author = post.find("span",style="text-transform: capitalize; font-size: 20px;font-weight: bold;color: #005681").get_text()   
    
#     print(post_url)
#     print(post_title)
#     print(post_author)
#     print("---")

    result.append([post_title, post_author, post_url])
    
```

```python
import csv

f = open("result.csv", "w")
writer= csv.writer(f)
writer.writerows(result)
f.close()
```
